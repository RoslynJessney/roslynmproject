```python
import numpy as np
from scipy import signal
from scipy.fft import fftshift
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 10, 0.1)
amplitude = np.sin(2*np.pi*2*t)
plt.plot(t, amplitude)
plt.title('Simple Sin Wave at 2Hz')
```

```python
fs=10 

f, t, Sxx = signal.spectrogram(amplitude, 10, nperseg=20)
plt.pcolormesh(t, f, Sxx, shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.colorbar()
plt.show()
```
